#### Here, we downsize the data from field_296 from 380571 rows to 100 rows. By assigning variable 'random' to field_296.sample, we are able to return a sample of items, in this case 100, from field_296.

```python
import random
random = field_296.sample(n=100, frac=None, replace=False, weights=None, random_state=1, axis=None, ignore_index=False)
random.to_csv(f'/Users/jillhughes/scope/random.csv', index=False)
random
```

#### Scatter plot created to show the dec classifier values vs the ra classifer values of the 100 randomized rows above. 

```python
ra = random.iloc[0:,4]
dec = random.iloc[0:,5]
plt.scatter(ra,dec)
plt.xlabel('ra Classifier',fontsize=14)
plt.ylabel('dec Classifier',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
```

#### Creates a histogram that shows the classification probability of a specific dnn and xgb classifier.

```python
#plots - histogram (class probs) - variable or periodic - xaxis 0-1
plt.hist(random['vnv_dnn'], bins = 10, range = (0, 1), label = 'vnv_dnn')
plt.hist(random['vnv_xgb'], bins = 10, range = (0, 1), alpha = 0.7, label = 'vnv_xgb')
plt.legend()
plt.xlabel('Classification Probability')
plt.ylabel('Number of Sources')
```
